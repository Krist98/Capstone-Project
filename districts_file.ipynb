```python
accomodates_list = []
for i in full_data.accomodates:
    k = re.findall("(\d+)\sguest", i)
    if k != []:
        accomodates_list.append(k)
    else:
        accomodates_list.append(['0'])
number = []   
accomodates_list = [item for sublist in accomodates_list for item in sublist]
for k in accomodates_list:
    a = int(k)
    number.append(a)
    
full_data.accomodates = number 
```

```python
for i in full_data.index:
    if "Shared half-bath" in  full_data.number_bathroom[i]:
        full_data.loc[i, 'number_bathroom'] = 0.5
    elif "Half-bath" in  full_data.number_bathroom[i]:
        full_data.loc[i, 'number_bathroom'] = 0.5
    elif "bath" not in  full_data.number_bathroom[i]:
        full_data.loc[i, 'number_bathroom'] = 0
    elif "private bath" in  full_data.number_bathroom[i]:
        t =  float(" ".join(re.findall("(\d+(?:\.\d+)?)\sprivate", full_data.number_bedroom[i])))
        full_data.loc[i, 'number_bathroom'] = t
    elif "shared bath" in  full_data.number_bathroom[i]:
        sh =  float(" ".join(re.findall("(\d+(?:\.\d+)?)\sshared\sbath", full_data.number_bedroom[i])))
        full_data.loc[i, 'number_bathroom'] = sh    
    else:
        
        k = re.findall("(\d+(?:\.\d+)?)\sbath", full_data.number_bedroom[i])
        k = "".join(k)
        if k != "":
            full_data.loc[i, 'number_bathroom'] = k 
```

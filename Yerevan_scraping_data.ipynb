```python
SLEEP = 7
 
data_hotel = pd.DataFrame({'city':[""], 'country':[""], 'host_since':[""], 'host_response_rate':[""], 'host_identity_verified':[""], 
                           'review':[""], 'superhost':[""],
                           'property_type':[""], 'accomodates':[""], 'number_bathroom':[""], 'number_bedroom':[""], 
                         'number_beds':[""], 'price':[""], 'cleanliness':[""], 'accuracy':[""], 'communication':[""], 
                           'location':[""], 'check_in':[""], 'value':[""]})


while True:
    links = driver.find_elements_by_class_name('m1v28t5c')
    time.sleep(1)
    for i in links:
        i.click()
        time.sleep(15)
        driver.switch_to_window(driver.window_handles[1]);
        some_element = driver.find_element_by_tag_name("html")
        some_element.send_keys(Keys.END)
        time.sleep(SLEEP)
        
        try: 
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('_9xiloll')).perform()
            time.sleep(1)
            city = driver.find_element_by_class_name('_9xiloll').text
            country = driver.find_element_by_class_name('_9xiloll').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('ciubx2o')).perform()
            time.sleep(1)
            cleanliness = driver.find_element_by_class_name('ciubx2o').text
            accuracy = driver.find_element_by_class_name('ciubx2o').text
            communication = driver.find_element_by_class_name('ciubx2o').text
            location = driver.find_element_by_class_name('ciubx2o').text
            check_in = driver.find_element_by_class_name('ciubx2o').text
            value = driver.find_element_by_class_name('ciubx2o').text
            time.sleep(1)


            ActionChains(driver).move_to_element(driver.find_element_by_class_name('s9fngse')).perform()
            time.sleep(1)
            host_since = driver.find_element_by_class_name('s9fngse').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('fhhmddr')).perform()
            time.sleep(1)
            host_response_rate = driver.find_element_by_class_name('fhhmddr').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('tq6hspd')).perform()
            time.sleep(1)
            host_identity_verified = driver.find_element_by_class_name('tq6hspd').text.split('\n')
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('tq6hspd')).perform()
            time.sleep(1)
            review = driver.find_element_by_class_name('tq6hspd').text.split('\n')
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('tq6hspd')).perform()
            time.sleep(1)
            superhost = driver.find_element_by_class_name('tq6hspd').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('_cv5qq4')).perform()
            time.sleep(1)
            Property_type = driver.find_element_by_class_name('_cv5qq4').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('_tqmy57')).perform()
            time.sleep(1)
            Accomodates = driver.find_element_by_class_name('_tqmy57').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('_tqmy57')).perform()
            time.sleep(1)
            number_bathroom = driver.find_element_by_class_name('_tqmy57').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('_tqmy57')).perform()
            time.sleep(1)
            number_bedroom = driver.find_element_by_class_name('_tqmy57').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('_tqmy57')).perform()
            time.sleep(1)
            number_beds = driver.find_element_by_class_name('_tqmy57').text
            ActionChains(driver).move_to_element(driver.find_element_by_class_name('_wgmchy')).perform()
            time.sleep(1)
            price = driver.find_element_by_class_name('_wgmchy').text
            data_hotel = data_hotel.append({'city':city, 'country': country,  'host_since': host_since, 'host_response_rate': host_response_rate, 
                                    'host_identity_verified': host_identity_verified, 'review': review, 'superhost': superhost,
                                    'property_type': Property_type, 'accomodates': Accomodates, 
                                    'number_bathroom': number_bathroom, 'number_bedroom': number_bedroom, 
                                    'number_beds': number_beds,  'price': price,
                                    'cleanliness': cleanliness, 'accuracy': accuracy, 'communication': communication, 
                       'location': location, 'check_in': check_in, 'value': value}, ignore_index=True)
        except:
            pass
            
        

        window_name = driver.window_handles[1]
        driver.switch_to_window(window_name = window_name)
        driver.close()
        driver.switch_to_window(driver.window_handles[0]);
        time.sleep(3)
    ActionChains(driver).move_to_element(driver.find_element_by_class_name('c1bdq4u1')).perform()
    if driver.find_element_by_class_name('c1bdq4u1').text == '15':
        break
    try:
        ActionChains(driver).move_to_element(driver.find_element_by_class_name('c1ytbx3a')).perform()
        driver.find_element_by_class_name('c1ytbx3a').click()
        time.sleep(7)
    except:
        pass
    
    
    
    
    
data_hotel.to_excel("Nork_Marash__data.xlsx", index = False)
data_hotel.to_csv("Nork_Marash__data.csv", index = False)
```
